```python
def T_iterative(n):
    T = [0] * (n + 1)
    T[0] = 5
    for i in range(1, n + 1):
        T[i] = 3 * T[i - 1] + 12 * i
    return T[n]

result = T_iterative(2)
print("T(2) =", result)
```

```python
#a. T(n) = T(n-1) + c
def solve_recursive_a(n, c):
    if n == 0:
        return 0
    else:
        return solve_recursive_a(n - 1, c) + c

c = 10  # value of constant c
result_a = solve_recursive_a(5, c)
print("a. T(5) =", result_a)
```

```python
# b. T(n) = 2T(n/2) + n
def solve_recursive_b(n):
    if n == 1:
        return 1
    else:
        return 2 * solve_recursive_b(n // 2) + n

result_b = solve_recursive_b(5)
print("b. T(5) =", result_b)
```

```python
# c. T(n) = 2T(n/2) + c
def solve_recursive_c(n, c):
    if n == 1:
        return c
    else:
        return 2 * solve_recursive_c(n // 2, c) + c

c = 10  # value of constant c
result_c = solve_recursive_c(5, c)
print("c. T(5) =", result_c)
```

```python
# d. T(n) = T(n/2) + c
def solve_recursive_d(n, c):
    if n == 1:
        return c
    else:
        return solve_recursive_d(n // 2, c) + c

c = 10  # value of constant c
result_d = solve_recursive_d(5, c)
print("d. T(5) =", result_d)

```

```python
import matplotlib.pyplot as plt
import networkx as nx

def draw_recursive_tree_a(n, level, parent, G, pos):
    if level == 0:
        return
    left_child = parent * 2
    right_child = parent * 2 + 1
    G.add_node(left_child, label=f"$T({n-1})$")
    G.add_node(right_child, label="1")
    G.add_edge(parent, left_child)
    G.add_edge(parent, right_child)
    pos[left_child] = (left_child, level - 1)
    pos[right_child] = (right_child, level - 1)
    draw_recursive_tree_a(n - 1, level - 1, left_child, G, pos)
    draw_recursive_tree_a(n - 1, level - 1, right_child, G, pos)

# Initialize the graph and position dictionary
G_a = nx.Graph()
pos_a = {}
root_a = 1
n = 4  # Define n before using it
G_a.add_node(root_a, label=f"$T({n})$")  # Corrected line
pos_a[root_a] = (root_a, 0)
# Draw the tree recursively
draw_recursive_tree_a(n, 3, root_a, G_a, pos_a)

# Draw the tree
plt.figure(figsize=(8, 6))
nx.draw(G_a, pos_a, with_labels=True, labels=nx.get_node_attributes(G_a, 'label'), node_size=1500, node_color='skyblue')
plt.title("Recursive Tree for T(n) = 2T(n-1) + 1")
plt.show()
```
